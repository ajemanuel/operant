```python
import numpy as np
import matplotlib.pyplot as plt
import compress_pickle as pickle
%matplotlib inline
```

```python
temp = pickle.load(r'E:/DATA/Behavior/20211124_122753_AGV5.gz')
```

```python
plt.figure(figsize=[5,20])
for trial in temp['di_data'].keys():
    if temp['results'][trial] == 'hit':
        plt.plot(temp['di_data'][trial]+trial,color='b')
    elif temp['results'][trial] == 'FA':
        plt.plot(temp['di_data'][trial]+trial,color='r')
    else:
        plt.plot(temp['di_data'][trial]+trial,color='gray')
        
plt.xlabel('Samples')
plt.ylabel('Trial')
```
